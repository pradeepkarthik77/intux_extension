```python
import os
import json
from pymongo import MongoClient

# MongoDB connection settings
mongo_uri = "mongodb://localhost:27017/"
database_name = "INTUX"
gazecollection = "GazeData"
clickcollection = "ClickData"
MetaCollection = "MetaData"

# Create a MongoDB client
client = MongoClient(mongo_uri)

# Select or create the database
db = client[database_name]

# Select or create the collection
collection = db[gazecollection]

# Directory containing your JSON files
json_files_directory = "E:\Mongodb_Backup\GazeDB"

# Iterate through each JSON file in the directory
for filename in os.listdir(json_files_directory):
    if filename.endswith(".json"):
        # Extract rollNo from the filename (assuming rollNo is the filename without extension)
        rollNo = os.path.splitext(filename)[0]

        # Load JSON data from the file
        with open(os.path.join(json_files_directory, filename), 'r') as file:
            data = json.load(file)

        # Add the extra field 'rollNo' to each document
        for document in data:
            document['rollNo'] = rollNo

        # Insert documents into MongoDB collection
        collection.insert_many(data)


# Select or create the collection
collection = db[clickcollection]

# Directory containing your JSON files
json_files_directory = "E:\Mongodb_Backup\ClickDB"

# Iterate through each JSON file in the directory
for filename in os.listdir(json_files_directory):
    if filename.endswith(".json"):
        # Extract rollNo from the filename (assuming rollNo is the filename without extension)
        rollNo = os.path.splitext(filename)[0]

        # Load JSON data from the file
        with open(os.path.join(json_files_directory, filename), 'r') as file:
            data = json.load(file)

        # Add the extra field 'rollNo' to each document
        for document in data:
            document['rollNo'] = rollNo

        # Insert documents into MongoDB collection
        collection.insert_many(data)

collection = db[MetaCollection]

# Directory containing your JSON files
json_files_directory = "E:\Mongodb_Backup\MetaDB"

# Iterate through each JSON file in the directory
for filename in os.listdir(json_files_directory):
    if filename.endswith(".json"):
        # Extract rollNo from the filename (assuming rollNo is the filename without extension)
        rollNo = os.path.splitext(filename)[0]

        # Load JSON data from the file
        with open(os.path.join(json_files_directory, filename), 'r') as file:
            data = json.load(file)

        # Insert documents into MongoDB collection
        collection.insert_many(data)

# Close the MongoDB connection
client.close()
```
